### Load the data

After exploring the file and folder structure, we have loaded the data for it's consumption.

```python
#load activities and set them to their meanings
activity_label = {int(key): value for key, value in eda_functions.get_key_value_pairs('../Data/UCI HAR Dataset/activity_labels.txt')}
activities = pd.read_table('../Data/UCI HAR Dataset/train/y_train.txt', header=None, delim_whitespace=True)
activities.columns = ['activity']
activities = activities.replace({'activity':activity_label})

#load subjects
subjects = [int(line.strip()) for line in open('../Data/UCI HAR Dataset/train/subject_train.txt','r')]

#load the feature names
features = np.array([line.strip().split(' ')[1] for line in open('../Data/UCI HAR Dataset/features.txt', 'r')])

#load the observations
observations= pd.read_table('../Data/UCI HAR Dataset/train/X_train.txt', header=None, delim_whitespace=True, names=features)
observations['subject'] = subjects
observations['activities'] = activities
```

### Histogram
For this, we will reduce the dataset to the total body acceleration columns, since they are the ones we have observed to have differences in their distribution.


```python
estimated = 'tBodyAcc'
meanfeatures = np.array([feature for feature in features if estimated in feature ])
meanfeatures = np.append(meanfeatures,['activities','subject'])
```

We now have plotted a histogram with the mean total body acceleration on the Z axis for all our subjects. Remember we expected it to have little variation.


```python
feature = 'tBodyAcc-mean()-Z'
maximums = observations[meanfeatures]

eda_functions.hist_plot(maximums[feature], feature, 100, True)
```

Now replicate the code for the maximum total body acceleration `tBodyAcc-max()-Z`

```python
feature = 'tBodyAcc-max()-Z'
maximums = observations[meanfeatures]

eda_functions.hist_plot(maximums[feature], feature, 50, True)
```

Now, we should take a look at only one subject, so we understand with one example what is going on. 
For this, subset the maximums data frame only have the subject == 1. `maximums[maximums['subject']==1]`, save it in a new dataset and replicate the previous code, first the mean and then the max values.

```python
subsetuser = maximums[maximums['subject']==1]
```

```python
plt.figure()
feature = 'tBodyAcc-mean()-Z'

eda_functions.hist_plot(maximums[feature], feature, 200, True)
```

With this, we can see that the mean differs. So there is variability between people.

We could argue that this may be because someone spent a longer time doing one activity rather than another.For this, le'ts use the `tBodyAcc-max()-Z` and look at when the subject is walking.

```python
plt.figure()
feature = 'tBodyAcc-max()-Z'

plt.hist(subsetuser.loc[subsetuser['activities']=='WALKING',feature], bins='auto')
plt.title(feature)
plt.show()
```

Now see what happens when he is sitting

```python
plt.figure()
feature = 'tBodyAcc-max()-Z'

plt.hist(subsetuser.loc[subsetuser['activities']=='SITTING',feature], bins='auto')
plt.title(feature)
plt.show()
```

We now know that there is variability.

- Activity variability: there seem to be some features that will give information to identify each activity.
- User variability: this seems to introduce some noise. Since as Data Scientist we should go from simple to more elaborate solutions, we will now only look at one user for our analysis until we are satisfied this ahs been enough.


If we start talking about its implementation, we may be implying that we want to create one model per user, instead of generalizing how any human being may perform each activity. In this scenario, the arrival of a new user, for whom there is no information to build their model. This is formally known as the **cold start** problem.

We shall not worry about that problem, for now, we are still understanding the data and it's behaviors and are not even sure we can generate something out of it. 

Once again: keep it simple.

### Scatterplot

So far we have been looking at one-dimensional analysis, and this is quite good.
never the less we have plenty of data and we want to see their interactions. A scatterplot is a bidimensional tool and happens to be useful for understanding correlations, data capacity among other things.

Since we have a lot of data, instead of plotting one combination of values, we can use `scatter_matrix` from the pandas plotting tools. This is particularly useful because it gives us the `density plot` in the diagonal. All we need to know about it is that the `density plot` is some sort of histogram smoothing.[NOTE 1].



We have simplified the plotting of all 3 axes of any variable by creating the plotting the `scattermatrixplot()` function. Further, we have already plotted the passive activities this way, so let's take a look.

```python
eda_functions.scattermatrixplot(subsetuser,'activities',['SITTING','STANDING','LAYING'],'tBodyAcc-max()')
```

It appears perfectly natural not to see any correlation in the total body acceleration for the resting activities since the person is not moving. Whereas the extra values may be related to the breathing movement or something else.

Now let's see if we can say a different story then the person is active. Please copy the code and look now for 'WALKING', 'WALKING_UPSTAIRS' and 'WALKING_DOWNSTAIRS'.

```python
eda_functions.scattermatrixplot(subsetuser,'activities',['WALKING','WALKING_UPSTAIRS','WALKING_DOWNSTAIRS'],'tBodyAcc-max()')
```
